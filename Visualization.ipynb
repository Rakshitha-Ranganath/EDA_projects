```python
# Data analysis and visualization
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data1=pd.read_csv('/BAC.csv')   # importing the data
data2=pd.read_csv('/C.csv')
data3=pd.read_csv('/GS.csv')
data4=pd.read_csv('/JPM.csv')
data5=pd.read_csv('/MS.csv')
```

```python
data1.head() # preview of data1
```

```python
bank_dataset = pd.concat([data1,data2,data3,data4,data5], axis=1, keys=['BAC', 'C', 'GS', 'JPM', 'MS']) # concatinate data of other table or bank
```

```python
bank_dataset.head()  # preview data frame
```

```python
sns.heatmap(data=bank_dataset.isnull())  # visalizing the null values using heat map
plt.title('Visualizing Null Values')
plt.show()
```

```python
sns.lineplot(data=bank_dataset)  # grapgh showing the visualization of returns
plt.title('Visualizing Returns')
plt.show()
```

```python
bank_dataset.xs('Close', level=1, axis=1).plot()  # this use to show the graph of the closing prices of banks over time
plt.title("Closing Prices of Banks Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
bank_dataset_volume = bank_dataset.xs('Volume', level=1, axis=1) # this use to show the graph of the trading volume of banks over time

bank_dataset_volume.plot(kind='area', stacked=True, alpha=0.4, title="Trading Volume of Banks Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend(title="Bank Symbols")
plt.grid(True)
plt.show()
```

```python
daily_return = bank_dataset.xs('Close', level=1, axis=1).pct_change() * 100  # this use to show the graph of the average daily return of each bank
avg_daily_return = daily_return.mean()
avg_daily_return.plot(kind='bar')
plt.title("Average Daily Return of Each Bank")
plt.xlabel("Bank names")
plt.ylabel("Average Daily Return (%)")
plt.show()
```
